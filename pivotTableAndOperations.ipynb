```python
import numpy as np
import pandas as pd
```

```python
data={
'Date': pd.date_range('2023-01-01', periods=20),
'Product': ['A', 'B', 'C', 'D'] * 5,
'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
           'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
'Sales': np.random.randint(100, 1000, 20),
'Units': np.random.randint(10, 100, 20),
'Rep': ['John', 'Mary', 'Bob', 'Alice', 'Bob', 'Alice', 'John', 'Mary','Bob',
        'John', 'Mary','John', 'Mary','Bob', 'Alice','Bob','John', 'Mary', 'Bob','Alice']
}

df = pd. DataFrame(data)
df ['Month'] = df ['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values='Sales',index='Region',columns='Product') # east with A ka mean
                                                                   # east not with A = NULL
```

```python
pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='median') #used to make clustered heat map
```

```python

pivot2= pd.pivot_table(df, values=['Sales', 'Units'], index='Region' ,columns='Product')
pivot2
```

#CrossTabs - counting purpose

```python
pd.crosstab(df['Region'],df['Product'])
```
