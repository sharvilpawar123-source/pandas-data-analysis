```python
import numpy as np
import pandas as pd
```

```python
data={
'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
'Date': pd.date_range('2023-01-01', periods=8)

}

df = pd. DataFrame(data)
df
```

```python
# gp by catagory and sum of sells
a =df.groupby('Category')
a
```

```python
for i,v in a:
  print(i)
  print(v)
```

```python
a =df.groupby('Category')['Sales'].sum()
a
```

```python
# gp by stores and calc sum of sales
b =df.groupby('Store')['Sales'].sum()
b
```

```python
# gp by multiple columns like gp by store and catagory
c =df.groupby(['Category','Store'])['Sales'].sum()
c

```
