# Finding Missing Data


```python
import numpy as np
import pandas as pd
data = {
'A': [1, 2, np.nan, 4, 5],
'B': [np.nan, 2, 3, 4,5],
'C': [1, 2, 3, np.nan, np.nan],
'D':[1, np.nan, np.nan, np.nan, np.nan]
}
df=pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()   #total null val in each row
```

```python
df.isna().any() # har ek row me NULL val he kya cheack karega
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3)
```

```python
# fill all null with 0
df.fillna(0)
```

```python
# # fill all null with diff vals
val={'A':0,'B':1,'C':2,'D':3}
df.fillna(value=val)
```

# merging ans sorting

```python
employees= pd.DataFrame({
'employee_id': [1, 2, 3, 4, 5],
'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
'department': ['HR', 'IT', 'Finance', 'IT','HR'],

})

# DataFrame 2: Salary information

salaries = pd.DataFrame({
'employee_id': [1, 2, 3, 6, 7],
'salary': [60000, 80000, 65000, 70000, 90000],
'bonus': [5000, 10000, 7000, 8000, 12000]})
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
# pd.merge(employees,salaries,on='employee_id',how='left')
# pd.merge(employees,salaries,on='employee_id',how='right')
# pd.merge(employees,salaries,on='employee_id',how='inner')
# pd.merge(employees,salaries,on='employee_id',how='cross')
```

#Concatination

```python
df1 = pd.DataFrame({

'A':['A0','A1','A2'],
'B':  ['B0','B1','B2'],
'C':['C0','C1','C2']

})

df2 = pd.DataFrame({

'A':['A3','A4','A5'],
'B':  ['B3','B4','B5'],
'C':['C3','C4','C5']

})

df1
```

```python
pd.concat([df1,df2])#on the basis of col
```

```python
pd.concat([df1,df2],axis=1)#on the basis of row
```

#joining two data frames

```python
df1 = pd.DataFrame({
'name':['Alice','Bob','Charlie']

}, index=[1, 2, 3])

# Second DataFrame

df2 = pd.DataFrame({'score': [85, 90, 75]

},index=[2,3, 4])
df1
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```
